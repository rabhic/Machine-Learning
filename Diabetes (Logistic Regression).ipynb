```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
from matplotlib import style
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

# Logistic Regression

```python
# Reading diabetes dataset
data_set = pd.read_csv('diabetes.csv')
data_set.head(7)

data_set.info()

# Splitting data into x and y part
x = data_set.iloc[:,0:8]
y = data_set.iloc[:,8]
type(x)

# obtaining train and test parts
x_train,y_train,x_test,y_test = train_test_split(x,y,test_size=0.30,random_state=100)
y_train

# Training model
log_model = LogisticRegression()
log_model.fit(x_train,x_test)

# Predicting y part with the help of trained model
y_predict = log_model.predict(y_train)

# Making dataframe for plotting
data_to_plot_log = pd.DataFrame({'Acutal ':y_test,'Predicted':y_predict})
data_to_plot_log.head(30)

# Printing Classification Report
print(classification_report(y_test,y_predict))

# Plotting confusion matrix
cm = confusion_matrix(y_test,y_predict)
cm
ax = plt.axes()
sns.heatmap(cm, annot=True, annot_kws={"size": 20},cmap="RdYlGn", )
ax.set_title('Confusion Matrix')
plt.show()

# Plotting ROC curve
pred_prob_log = log_model.predict_proba(y_train)
fpr_log, tpr_log, thresh_log = roc_curve(y_test, pred_prob_log[:,1], pos_label=1)
auc_score_log = roc_auc_score(y_test, pred_prob_log[:,1])
print(auc_score_log)


plt.title('Receiver Operating Characteristic')
plt.plot(fpr_log, tpr_log, 'b', label = 'AUC = %0.2f' )
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
