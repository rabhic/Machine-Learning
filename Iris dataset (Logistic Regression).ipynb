# Objective :- Write a program in python using Logistic Regression to classify Iris Dataset into different Classes.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
from matplotlib import style
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve


# Loading Iris dataset
iris_data_set = sns.load_dataset('iris')

# Printing info about Iris dataset
load_dataset= iris_data_set.info()

# Getting unique classes (flowers) in Iris dataset
iris_data_set['species'].unique()

# Getting all different flowers in different variables.
setosa_data = iris_data_set[iris_data_set['species']=='setosa']
versicolor_data = iris_data_set[iris_data_set['species']=='versicolor']
verginica_data = iris_data_set[iris_data_set['species']=='virginica']

# Plotting sepal length vs sepal width for setosa class
plt.plot(setosa_data['sepal_length'],setosa_data['sepal_width'],'og')
plt.title('Sepal length vs Sepal Width (Setosa)',size = 20)
plt.xlabel('sepal length',size=15)
plt.ylabel('sepal width',size=15)
plt.show()

# Plotting sepal length vs sepal width for verginica class
plt.plot(verginica_data['sepal_length'],verginica_data['sepal_width'],'ob')
plt.title('Sepal length vs Sepal Width (Verginica)',size = 20)
plt.xlabel('sepal length',size=15)
plt.ylabel('sepal width',size=15)
plt.show()

# Plotting sepal length vs sepal width for veriscolor class
plt.plot(versicolor_data['sepal_length'],versicolor_data['sepal_width'],'or')
plt.title('Sepal length vs Sepal Width (Versicolor)',size = 20)
plt.xlabel('sepal length',size=15)
plt.ylabel('sepal width',size=15)
plt.show()

# Plotting petal length vs petal width for setosa class
plt.plot(setosa_data['petal_length'],setosa_data['petal_width'],'og')
plt.title('Petal length vs Petal Width (setosa)',size = 20)
plt.xlabel('Petal length',size=15)
plt.ylabel('Petal width',size=15)
plt.show()

# Plotting petal length vs petal width for verginica class
plt.plot(verginica_data['petal_length'],verginica_data['petal_width'],'ob')
plt.title('Petal length vs Petal Width (Verginica)',size = 20)
plt.xlabel('Petal length',size=15)
plt.ylabel('Petal width',size=15)
plt.show()

# Plotting petal length vs petal width for versicolor class
plt.plot(versicolor_data['petal_length'],versicolor_data['petal_width'],'or')
plt.title('Petal length vs Petal Width (Versicolor)',size = 20)
plt.xlabel('Petal length',size=15)
plt.ylabel('Petal width',size=15)
plt.show()

# Getting features in variable x and label in variable y
x = iris_data_set.drop('species',axis=1)
y= iris_data_set['species']

# Printing top 7 rows of varibale x (features)
x.head(7)

# Printing top 7 rows of variable y (labels/class/target)
y.head(7)

# Using tarin test split function to split our dataset into train and test parts
x_train,y_train,x_test,y_test = train_test_split(x,y,test_size=.25,random_state=150)

# Printing rows of x_train
x_train

# Printing rows of x_test
x_test

# Printing rows of y_train
y_train

# Printing rows of y_test
y_test

# Creating object for LogisticRegression class which will be our model
model = LogisticRegression()

# Using fit function we are training our model
model.fit(x_train,x_test)

# Obtaining Predicted value for y_train and printing it
y_predict = model.predict(y_train)
y_predict

# Obtaining dataframe for avtual vs predicted value for testing part and ploting 
data_to_plot = pd.DataFrame({'Actual ': y_test,'Predicted':y_predict})
data_to_plot

# Printing Classification report for our classification model (using Logistic Regression)
print(classification_report(y_test, y_predict))
cm = confusion_matrix(y_test,y_predict)
cm

# Plotting heatmap graph to show confusion matrix
ax = plt.axes()
df_cm = cm
sns.heatmap(df_cm, annot=True, annot_kws={"size": 40},cmap="RdYlGn", )
ax.set_title('Confusion Matrix')
plt.show()
```
